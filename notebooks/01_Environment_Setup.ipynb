```python
import os
import sys

def running_in_colab():
    try:
        import google.colab
        return True
    except ImportError:
        return False

IS_COLAB = running_in_colab()

print("=" * 60)
print("🐟 AquaTutor Environment Detection")
print("=" * 60)

if IS_COLAB:
    print("Environment : Google Colab")
else:
    print("Environment : Local Jupyter / Cursor")

print("=" * 60)
```

```python
import importlib

required_packages = {
    "fiftyone": "fiftyone",
    "ultralytics": "ultralytics",
    "torch": "torch",
    "PIL": "Pillow"
}

missing = []

print("Checking packages...\n")

for module, package in required_packages.items():
    try:
        importlib.import_module(module)
        print(f"✅ {package}")
    except ImportError:
        print(f"❌ {package}")
        missing.append(package)
```

```python
if missing:
    print("\nInstalling missing packages...")

    packages = " ".join(missing)

    !pip install --upgrade {packages} --quiet

    install_required = True

else:

    print("\nAll packages already installed.")

    install_required = False
```

```python
print("=" * 60)
print("Environment Ready")
print("=" * 60)

print(f"Running in Colab : {IS_COLAB}")
print(f"Packages Missing : {len(missing)}")
print(f"Installation Run : {install_required}")

print("=" * 60)
```
