```python
# detect environment

import os
import sys
import platform
from pathlib import Path

try:
    import google.colab
    IS_COLAB = True
except ImportError:
    IS_COLAB = False

print("Environment:", "Google Colab" if IS_COLAB else "Local / Cursor / Jupyter")
print("Current directory:", os.getcwd())
```

```python
# Clone or Locate Repository

REPO_URL = "https://github.com/johnnyshorten/Aquatutor.git"
REPO_NAME = "Aquatutor"

def find_project_root(start_path=None):
    """
    Walk upwards from the current directory until a .git folder is found.
    """
    current = Path(start_path or Path.cwd()).resolve()

    for path in [current] + list(current.parents):
        if (path / ".git").exists():
            return path

    return None


if IS_COLAB:
    repo_path = Path("/content") / REPO_NAME

    if not repo_path.exists():
        print("Cloning AquaTutor repository...")
        !git clone {REPO_URL}

    os.chdir(repo_path)

else:
    project_root = find_project_root()

    if project_root is None:
        raise FileNotFoundError(
            "AquaTutor project root not found. "
            "Open this notebook from inside the local Aquatutor repository."
        )

    os.chdir(project_root)

PROJECT_ROOT = Path.cwd()

print("Project root:", PROJECT_ROOT)
```

```python
# import 
package_status = {}

try:
    import torch
    package_status["torch"] = True
except Exception as e:
    package_status["torch"] = False
    print("Torch import failed:", e)

try:
    import fiftyone as fo
    import fiftyone.zoo as foz
    package_status["fiftyone"] = True
except Exception as e:
    package_status["fiftyone"] = False
    print("FiftyOne import failed:", e)

try:
    import ultralytics
    package_status["ultralytics"] = True
except Exception as e:
    package_status["ultralytics"] = False
    print("Ultralytics import failed:", e)
```

```python
# Verify project structire

required_folders = [
    "assets",
    "datasets",
    "datasets/raw",
    "datasets/processed",
    "datasets/exports",
    "docs",
    "docs/handbook",
    "experiments",
    "figures",
    "journal",
    "models",
    "notebooks",
    "presentations",
    "reports",
    "results",
    "scripts",
    "src",
]

missing_folders = []

print("Checking project folders...\n")

for folder in required_folders:
    folder_path = PROJECT_ROOT / folder

    if folder_path.exists():
        print(f"✅ {folder}")
    else:
        print(f"❌ {folder}")
        missing_folders.append(folder)
```

```python
# Verify Project Files

required_files = [
    "README.md",
    "LICENSE",
    "CHANGELOG.md",
    "CONTRIBUTING.md",
    ".gitignore",
    "notebooks/01_Environment_Setup.ipynb",
    "notebooks/02_Run_Environment.ipynb",
]

missing_files = []

print("Checking project files...\n")

for file in required_files:
    file_path = PROJECT_ROOT / file

    if file_path.exists():
        print(f"✅ {file}")
    else:
        print(f"❌ {file}")
        missing_files.append(file)
```

```python
# Hardware Check

gpu_available = False
gpu_name = "CPU only"

if package_status.get("torch", False):
    gpu_available = torch.cuda.is_available()

    if gpu_available:
        gpu_name = torch.cuda.get_device_name(0)

print("GPU Available:", gpu_available)
print("GPU:", gpu_name)
```

```python
# Package versions

print("Package versions:\n")

if package_status.get("torch", False):
    print("Torch       :", torch.__version__)
else:
    print("Torch       : Not available")

if package_status.get("fiftyone", False):
    print("FiftyOne    :", fo.__version__)
else:
    print("FiftyOne    : Not available")

if package_status.get("ultralytics", False):
    print("Ultralytics :", ultralytics.__version__)
else:
    print("Ultralytics : Not available")
```

```python
# Run FiftyOne 

RUN_FIFTYONE_TEST = False

if RUN_FIFTYONE_TEST:

    if package_status.get("fiftyone", False):

        dataset = foz.load_zoo_dataset(
            "quickstart",
            max_samples=20
        )

        session = fo.launch_app(dataset, auto=False)
        session.open_tab()

    else:
        print("FiftyOne is not available.")

else:
    print("FiftyOne Skiped")
```

```python
print("\n" + "=" * 70)
print("🐟 AquaTutor Environment Status")
print("=" * 70)

print(f"Environment        : {'Google Colab' if IS_COLAB else 'Local / Cursor / Jupyter'}")
print(f"Project Root       : {PROJECT_ROOT}")
print(f"Python Version     : {platform.python_version()}")
print(f"Platform           : {platform.platform()}")
print()

print("Repository")
print(f"Git Repository     : {'✅ Found' if (PROJECT_ROOT / '.git').exists() else '❌ Missing'}")
print(f"Required Folders   : {'✅ OK' if not missing_folders else '❌ Missing'}")
print(f"Required Files     : {'✅ OK' if not missing_files else '❌ Missing'}")
print()

print("Hardware")
print(f"GPU Available      : {'✅ YES' if gpu_available else '❌ NO'}")
print(f"GPU Device         : {gpu_name}")
print()

print("Packages")
print(f"Torch              : {'✅ OK' if package_status.get('torch') else '❌ Missing'}")
print(f"FiftyOne           : {'✅ OK' if package_status.get('fiftyone') else '❌ Missing'}")
print(f"Ultralytics        : {'✅ OK' if package_status.get('ultralytics') else '❌ Missing'}")
print()
print(f"FiftyOne      : {'✅ Not Skipped' if RUN_FIFTYONE_TEST else '❌ Skipped'}")

if missing_folders:
    print("Missing folders:")
    for folder in missing_folders:
        print(f" - {folder}")
    print()

if missing_files:
    print("Missing files:")
    for file in missing_files:
        print(f" - {file}")
    print()

ready = (
    not missing_folders
    and package_status.get("torch", False)
    and package_status.get("fiftyone", False)
    and package_status.get("ultralytics", False)
)

print("=" * 70)

if ready:
    print("✅ Environment Ready")
    print("Next Notebook: 03_Dataset_Selection.ipynb")
else:
    print("⚠️ Environment has issues to fix before continuing.")

print("=" * 70)
```
